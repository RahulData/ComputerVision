```python
import os
from PIL import Image
import cv2
import matplotlib.pyplot as plt
# print(os.getcwd())
# PIL.Image('faces.jpg')

pil_im = Image.open('faces.jpg')
#load test iamge
test1 = cv2.imread('faces.jpg')
#convert the test image to gray image as opencv face detector expects gray images 
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
 #load test iamge
# gray_img.shape (480,720)
plt.figure(figsize=(9,6))
plt.imshow(gray_img,cmap='gray')
plt.show()
```

```python
#load cascade classifier training file for haarcascade 
haar_face_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_frontalface_alt.xml');
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5);  

haar_eyes_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_eye.xml');
eyes = haar_eyes_cascade.detectMultiScale(gray_img, scaleFactor=1.03, minNeighbors=5);  

#print the number of faces found 
print('Faces found: ', len(faces))
print('Eyes found: ', len(eyes))


#print the number of faces found 
# print('Faces found: ', len(faces))
```

```python
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#go over list of faces and draw them as rectangles on original colored img 
for (x, y, w, h) in faces:     
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 255), 2)
#convert image to RGB and show image 
plt.figure(figsize=(9,6))
plt.imshow(convertToRGB(test1))
plt.show()
```

```python
test = cv2.imread('faces.jpg')
test1 = test.copy()
for (x, y, w, h) in eyes:     
    cv2.rectangle(test1, (x, y), (x+w, y+h), (255, 0, 0), 2)    
plt.figure(figsize=(9,6))
plt.imshow(convertToRGB(test1))
plt.show()
```

```python
test = cv2.imread('faces.jpg')
test1 = test.copy()
for (x, y, w, h) in faces:     
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 255), 2)
for (x, y, w, h) in eyes:     
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 0, 255), 2)    
plt.figure(figsize=(9,6))
plt.imshow(convertToRGB(test1))
plt.show()
```
